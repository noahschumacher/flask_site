```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import pickle
```

```python
df = pd.read_csv('cars.csv')
X = df[['cylinders', 'horsepower', 'weight']]
target = df.mpg
model = LinearRegression()
model.fit(X, target)
print(model.coef_)
```

```python
pickle.dump(model, open('linreg.p', 'wb'))
```

```python
import pickle
model = pickle.load(open('linreg.p', 'rb'))
```
